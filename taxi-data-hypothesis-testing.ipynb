```python
# Hipóteses:
# H0 (nula): μ_sábados_bad = μ_sábados_good
# H1 (alternativa): μ_sábados_bad ≠ μ_sábados_good

df = pd.read_csv('/datasets/project_sql_result_07.csv')

# PRIMEIRO: converter start_ts para datetime
df['start_ts'] = pd.to_datetime(df['start_ts'])

# AGORA pode usar .dt.day_name()
df['weekday'] = df['start_ts'].dt.day_name()

df_saturday = df[df['weekday'] == 'Saturday']

bad_weather = df_saturday[df_saturday['weather_conditions'] == 'Bad']['duration_seconds'].dropna()
good_weather = df_saturday[df_saturday['weather_conditions'] == 'Good']['duration_seconds'].dropna()

print("Corridas em sábados com tempo ruim:", len(bad_weather))
print("Corridas em sábados com tempo bom:", len(good_weather))

alpha = 0.05
t_stat, p_val = stats.ttest_ind(bad_weather, good_weather, equal_var=False)
print("t-stat:", t_stat)
print("p-value:", p_val)

if p_val < alpha:
    print("Rejeitamos H0: há diferença significativa na duração média das corridas.")
else:
    print("Não rejeitamos H0: não há evidência de diferença significativa.")
```

A hipótese nula assume que a duração média das corridas do Loop para O’Hare não muda nos sábados chuvosos.
A hipótese alternativa considera que essa duração média muda nesses dias.
Definimos o nível de significância em 5%.
Compararam-se duas amostras independentes: corridas em sábados chuvosos e corridas nos demais dias.
O critério escolhido foi o teste t de Student para duas amostras independentes.
Esse teste é adequado porque compara médias de variáveis contínuas entre grupos distintos.


Conclusão:
O teste mostrou p‑valor ≈ 6.7e‑12, muito menor que 0.05. Portanto, rejeitamos H₀ e concluímos que a duração média das corridas aos sábados difere significativamente entre tempo bom e ruim.
